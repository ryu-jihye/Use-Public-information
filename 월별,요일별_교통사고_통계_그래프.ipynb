```
import seaborn as sns
```

```
plt.figure(figsize=(10, 3))
sns.barplot(data=df_last, x="사고건수", y="월", ci=None)
```

```
sns.barplot(data=df_last, x="요일", y="사고건수")
```

```
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_last, x="월", y="사고건수", hue="요일")
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)
```

```
sns.relplot(data=df_last, x="월", y="사고건수", hue="요일", kind="line", col="요일", col_wrap=4, ci=None)
```

```
sns.catplot(data=df_last, x="월", y="사고건수", kind="bar", col="요일", col_wrap=4)
```

```
sns.boxplot(data=df_last, x="월", y="사고건수")
```

```
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_last, x="월", y="사고건수", hue="요일")
```

```
sns.violinplot(data=df_last, x="월", y="사고건수")
```

```
plt.figure(figsize=(15, 3))
sns.swarmplot(data=df_last, x="월", y="사고건수", hue="요일")
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)
```
