```
m = folium.Map(location=[lat, long], zoom_start=12)
folium.Marker([33.2514, 126.428], tooltip="탑피자").add_to(m)
```

```
for i in af_m.index[:10]:
    tooltip = af_m.loc[i, "상호명"] +"-"+ af_m.loc[i, "도로명주소"]
    lat = af_m.loc[i, "위도"]
    long = af_m.loc[i, "경도"]
    
    print(tooltip, lat, long)
```

```
m = folium.Map(location=[lat, long], zoom_start=12, 
    tiles='Stamen Toner')

for i in af_m.index[:100]:
    tooltip = af_m.loc[i, "상호명"] +"-"+ af_m.loc[i, "도로명주소"]
    lat = af_m.loc[i, "위도"]
    long = af_m.loc[i, "경도"]
    
    folium.CircleMarker([lat, long], tooltip=tooltip, radius=3).add_to(m)
```
